```python
import numpy as np
import matplotlib.pyplot as plt

n = np.random.randint(1, 101)
y = []
for i in range(n):
  y_i = np.random.randint(0, 2) #Generando exitos y fracasos
  y.append(y_i)
n1 = np.sum(y)
n0 = n - n1
print("n1 = exitos = ", n1)
print("n2 = fracasos = ", n0)
print(y)
```

```python
alpha = 2
beta = 8
alpha_post = alpha + n1
beta_post = beta + n0

print("alpha_post = ", alpha_post)
print("beta_post = ", beta_post)
```

```python
p1_bayes = alpha_post / (alpha_post + beta_post)
p0_bayes = 1 - p1_bayes
print("p1_bayes = ", p1_bayes)
print("p0_bayes = ", p0_bayes)
```

```python
M = 20 #Observar Knew en M ensayos
R = 10000

S_samples=[]

for r in range(R):
    theta_r = np.random.beta(alpha_post, beta_post)
    S_r = np.random.binomial(M, theta_r) #Utilizando un plugin no hay variación sobre theta_r
    S_samples.append(S_r)

```

```python
S_samples = np.array(S_samples)
```

```python
plt.bar(np.arange(M+1), np.bincount(S_samples, minlength=M+1))
plt.xlabel("Numero de éxitos futuros")
plt.ylabel("Probabilidad")
plt.title("Predictiva usando Bayes")
plt.show()
```

```python
values, counts = np.unique(S_samples, return_counts=True)
rel_freq = counts / R
print("Values:", values)
print("Relative Frequencies:", rel_freq)
print(np.sum(rel_freq))
print(np.unique(S_samples, return_counts=True))
```

```python
if alpha_post > 1 and beta_post >1:
  theta_map = (alpha_post-1)/(alpha_post + beta_post -2)
else:
  tehata_map = alpha_post/ (alpha_post + beta_post)
print("theta_map = ", theta_map)
```

```python
S_plugin = np.random.binomial(M, theta_map, size = R)
print("S_plugin = ", S_plugin)
```

```python
plt.bar(np.arange(M + 1), np.bincount(S_plugin, minlength=M + 1))
plt.xlabel("Numero de éxitos futuros")
plt.ylabel("Probabilidad")
plt.title("Predictiva usando Plug_in")
plt.show()
```
